# Silver Layer Design - Clinical Trial Lakehouse

## Objective
Transform Bronze Parquet data into clean, conformed Delta tables stored in ADLS Gen2 Silver layer.

## Scope
- 18 clinical trial tables
- No joins in Silver
- Standardization, data quality, and type normalization only

## Output
- Delta tables in abfss://silver@hishamclinicaldl.dfs.core.windows.net/<table_name>

# Silver Layer Principles

1. Preserve original business entities (no joins)
2. Standardize data types (especially decimals and timestamps)
3. Add audit columns for traceability
4. Ensure schema consistency across all tables
5. Maintain one Delta table per source table

# Silver Transformation Rules (Core Logic)
## Standardization, Audit Columns, and Decimal Normalization

```python
from pyspark.sql.functions import col
from pyspark.sql.types import DecimalType

def normalize_decimals(df):

    for field in df.schema.fields:
        if str(field.dataType).startswith("DecimalType"):

            df = df.withColumn(
                field.name,
                col(field.name).cast(DecimalType(12, 2))
            )

    return df
```

# Silver Transformation Engine v3
## Full pipeline: audit + normalization

```python
from pyspark.sql.functions import current_timestamp, lit, col
from pyspark.sql.types import DecimalType

def normalize_decimals(df):

    for field in df.schema.fields:
        if str(field.dataType).startswith("DecimalType"):
            df = df.withColumn(
                field.name,
                col(field.name).cast(DecimalType(12, 2))
            )

    return df


def transform_silver(df, table_name):

    # audit columns
    df = df.withColumn("silver_loaded_at", current_timestamp()) \
           .withColumn("source_table", lit(table_name))

    # normalize decimals
    df = normalize_decimals(df)

    return df
```

# Silver Load Execution
## Applying transformation engine across all tables

```python
for table in tables:

    print(f"Processing Silver table: {table}")

    bronze_path = f"{BRONZE_PATH}{table}/"
    silver_path = f"{SILVER_PATH}{table}/"

    # Read Bronze
    df = spark.read.parquet(bronze_path)

    # Transform
    df_silver = transform_silver(df, table)

    # Write to Silver (Delta)
    df_silver.write \
        .format("delta") \
        .mode("overwrite") \
        .save(silver_path)

    print(f"Completed: {table}")
```

```python
%sql
CREATE OR REPLACE TABLE databricks_clinical_trial.gold.patient_study_summary
USING DELTA
LOCATION 'abfss://gold@hishamclinicaldl.dfs.core.windows.net/patient_study_summary/'
AS
SELECT
    p.patient_id,
    p.patient_code,
    p.gender,
    p.race,
    p.ethnicity,
    p.date_of_birth,
    floor(datediff(current_date(), p.date_of_birth) / 365.25) AS age_years,

    e.enrollment_id,
    e.enrollment_date,
    e.status AS enrollment_status,

    s.study_id,
    s.study_code,
    s.study_title,
    s.therapeutic_area,
    s.phase,

    a.arm_id,
    a.arm_name,

    r.site_id,
    r.site_name
FROM databricks_clinical_trial.silver.patients p
INNER JOIN databricks_clinical_trial.silver.enrollments e
    ON p.patient_id = e.patient_id
INNER JOIN databricks_clinical_trial.silver.clinical_studies s
    ON e.study_id = s.study_id
LEFT JOIN databricks_clinical_trial.silver.study_arms a
    ON e.arm_id = a.arm_id
LEFT JOIN databricks_clinical_trial.silver.research_sites r
    ON e.site_id = r.site_id;
```

```python
%sql
CREATE OR REPLACE TABLE databricks_clinical_trial.gold.lab_summary
USING DELTA
LOCATION 'abfss://gold@hishamclinicaldl.dfs.core.windows.net/lab_summary/'
AS
SELECT
    l.lab_result_id,
    l.visit_id,
    l.patient_id,

    p.patient_code,
    p.gender,
    floor(datediff(current_date(), p.date_of_birth) / 365.25) AS age_years,

    v.enrollment_id,
    e.study_id,
    e.site_id,

    l.lab_test_name,
    l.lab_test_code,
    l.lab_panel,
    l.result_value,
    l.result_unit,
    l.normal_range_low,
    l.normal_range_high,
    l.is_abnormal,
    l.toxicity_grade,

    CASE WHEN l.is_abnormal = true THEN 'Abnormal' ELSE 'Normal' END AS lab_status,

    CASE
        WHEN l.toxicity_grade >= 3 THEN 'High Risk'
        WHEN l.toxicity_grade = 2 THEN 'Moderate Risk'
        ELSE 'Low Risk'
    END AS toxicity_risk_category,

    l.collection_datetime,
    l.result_datetime
FROM databricks_clinical_trial.silver.lab_results l
LEFT JOIN databricks_clinical_trial.silver.patients p
    ON l.patient_id = p.patient_id
LEFT JOIN databricks_clinical_trial.silver.clinical_visits v
    ON l.visit_id = v.visit_id
LEFT JOIN databricks_clinical_trial.silver.enrollments e
    ON v.enrollment_id = e.enrollment_id;
```

```python
%sql
CREATE OR REPLACE TABLE databricks_clinical_trial.gold.safety_adverse_events
USING DELTA
LOCATION 'abfss://gold@hishamclinicaldl.dfs.core.windows.net/safety_adverse_events/'
AS
SELECT
    ae.ae_id,
    ae.patient_id,
    ae.visit_id,
    ae.enrollment_id,
    ae.reporter_investigator_id,

    p.patient_code,
    p.gender,
    floor(datediff(current_date(), p.date_of_birth) / 365.25) AS age_years,

    e.study_id,
    e.site_id,

    ae.ae_term,
    ae.meddra_pt_code,
    ae.severity,
    ae.ctcae_grade,
    ae.onset_date,
    ae.resolution_date,
    ae.outcome,
    ae.is_serious,
    ae.sae_category,
    ae.causality,
    ae.action_taken,
    ae.is_expected,

    CASE
        WHEN ae.is_serious = true THEN 'Serious AE'
        WHEN ae.ctcae_grade >= 3 THEN 'High Grade AE'
        WHEN ae.ctcae_grade = 2 THEN 'Moderate AE'
        ELSE 'Low Grade AE'
    END AS ae_risk_category,

    CASE WHEN ae.resolution_date IS NULL THEN 'Ongoing' ELSE 'Resolved' END AS ae_status,
    datediff(ae.resolution_date, ae.onset_date) AS ae_duration_days,

    ae.created_at,
    ae.updated_at
FROM databricks_clinical_trial.silver.adverse_events ae
LEFT JOIN databricks_clinical_trial.silver.patients p
    ON ae.patient_id = p.patient_id
LEFT JOIN databricks_clinical_trial.silver.enrollments e
    ON ae.enrollment_id = e.enrollment_id;
```

```python
%sql
CREATE OR REPLACE TABLE databricks_clinical_trial.gold.visit_summary
USING DELTA
LOCATION 'abfss://gold@hishamclinicaldl.dfs.core.windows.net/visit_summary/'
AS
SELECT
    v.visit_id,
    v.enrollment_id,
    v.investigator_id,
    v.visit_number,
    v.visit_name,
    v.visit_type,

    e.patient_id,
    p.patient_code,
    e.study_id,
    e.site_id,

    v.scheduled_date,
    v.actual_visit_date,
    v.visit_window_start,
    v.visit_window_end,
    v.visit_status,

    CASE
        WHEN v.actual_visit_date IS NULL THEN 'Not Completed'
        WHEN v.actual_visit_date BETWEEN v.visit_window_start AND v.visit_window_end THEN 'On Time'
        WHEN v.actual_visit_date < v.visit_window_start THEN 'Early'
        WHEN v.actual_visit_date > v.visit_window_end THEN 'Late'
        ELSE 'Unknown'
    END AS visit_timing_status,

    datediff(v.actual_visit_date, v.scheduled_date) AS days_from_scheduled,

    v.notes,
    v.created_at,
    v.updated_at
FROM databricks_clinical_trial.silver.clinical_visits v
LEFT JOIN databricks_clinical_trial.silver.enrollments e
    ON v.enrollment_id = e.enrollment_id
LEFT JOIN databricks_clinical_trial.silver.patients p
    ON e.patient_id = p.patient_id;
```
