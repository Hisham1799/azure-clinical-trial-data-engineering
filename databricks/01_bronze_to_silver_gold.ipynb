## Step 1 Ã¢â‚¬â€ Infrastructure Setup
### Register External Locations in Unity Catalog
Points Unity Catalog to the Bronze, Silver, and Gold containers in ADLS Gen2 using the `cred-clinical-adls` storage credential (Azure Managed Identity). This must run once before any data reads or writes.

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `clinical-trial-adls`
URL 'abfss://bronze@hishamclinicaldl.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `cred-clinical-adls`)
COMMENT 'Bronze layer - raw Parquet ingested via ADF';

CREATE EXTERNAL LOCATION IF NOT EXISTS `clinical-trial-silver`
URL 'abfss://silver@hishamclinicaldl.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `cred-clinical-adls`)
COMMENT 'Silver layer - cleaned Delta tables';

CREATE EXTERNAL LOCATION IF NOT EXISTS `clinical-trial-gold`
URL 'abfss://gold@hishamclinicaldl.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `cred-clinical-adls`)
COMMENT 'Gold layer - business-ready Delta tables';
```

## Step 2 Ã¢â‚¬â€ Unity Catalog Schema Setup
Creates the Silver and Gold schemas inside the `databricks_clinical_trial` catalog. Schemas are like folders inside the catalog Ã¢â‚¬â€ they group related tables together.

```python
%sql
CREATE SCHEMA IF NOT EXISTS databricks_clinical_trial.silver;
CREATE SCHEMA IF NOT EXISTS databricks_clinical_trial.gold;
```

## Step 3 Ã¢â‚¬â€ Bronze Layer Exploration
### Schema Discovery Across All 18 Tables
Reads each Bronze Parquet table directly from ADLS Gen2 and prints the row count and column schema. Bronze is the raw, immutable copy of the source MySQL data Ã¢â‚¬â€ no transformations applied.

```python
tables = [
    "adverse_events", "clinical_studies", "clinical_visits",
    "concomitant_medications", "dropouts", "drug_administration",
    "enrollments", "investigators", "lab_results", "medical_history",
    "medications", "patients", "protocol_deviations", "questionnaires",
    "research_sites", "study_arms", "study_sites", "vital_signs"
]

for table in tables:
    print("=" * 80)
    print(f"TABLE: {table}")
    df = spark.read.parquet(f"abfss://bronze@hishamclinicaldl.dfs.core.windows.net/{table}/")
    print(f"Rows: {df.count()}")
    df.printSchema()
    print()
```

## Step 4 Ã¢â‚¬â€ Silver Layer Transformation
### Design Principles
1. One Delta table per source table Ã¢â‚¬â€ no joins in Silver
2. Normalize excessive decimal precision (`decimal(38,18)` Ã¢â€ â€™ `decimal(12,2)`)
3. Add audit columns: `silver_loaded_at` (timestamp) and `source_table` (string)
4. Preserve all original columns Ã¢â‚¬â€ cleaning only, no business logic
5. Write as Delta format for ACID transactions and time travel

```python
from pyspark.sql.functions import current_timestamp, lit, col
from pyspark.sql.types import DecimalType

BRONZE_PATH = "abfss://bronze@hishamclinicaldl.dfs.core.windows.net/"
SILVER_PATH = "abfss://silver@hishamclinicaldl.dfs.core.windows.net/"

def normalize_decimals(df):
    """Cast all decimal(38,18) columns to decimal(12,2) for storage efficiency."""
    for field in df.schema.fields:
        if str(field.dataType).startswith("DecimalType"):
            df = df.withColumn(field.name, col(field.name).cast(DecimalType(12, 2)))
    return df

def transform_silver(df, table_name):
    """Apply Silver transformations: audit columns + decimal normalization."""
    df = df.withColumn("silver_loaded_at", current_timestamp()) \
           .withColumn("source_table", lit(table_name))
    df = normalize_decimals(df)
    return df
```

### Silver Load Execution
Runs the transformation function across all 18 tables. Reads each table from Bronze (Parquet), applies cleaning, writes as Delta to the Silver container. Mode is `overwrite` Ã¢â‚¬â€ safe to re-run.

```python
for table in tables:
    print(f"Processing: {table}")
    df = spark.read.parquet(f"{BRONZE_PATH}{table}/")
    df_silver = transform_silver(df, table)
    df_silver.write \
        .format("delta") \
        .mode("overwrite") \
        .save(f"{SILVER_PATH}{table}/")
    print(f"  Done: {table}")

print("\nAll 18 Silver tables written successfully.")
```

## Step 5 Ã¢â‚¬â€ Register Silver Tables in Unity Catalog
Delta files now exist in ADLS. This step registers each file path as a named external table in Unity Catalog so SQL queries can reference them by three-part name (`catalog.schema.table`). No data is moved Ã¢â‚¬â€ this is metadata registration only.

```python
silver_tables = [
    "adverse_events", "clinical_studies", "clinical_visits",
    "concomitant_medications", "dropouts", "drug_administration",
    "enrollments", "investigators", "lab_results", "medical_history",
    "medications", "patients", "protocol_deviations", "questionnaires",
    "research_sites", "study_arms", "study_sites", "vital_signs"
]

for t in silver_tables:
    spark.sql(f"""
        CREATE TABLE IF NOT EXISTS databricks_clinical_trial.silver.{t}
        USING DELTA
        LOCATION 'abfss://silver@hishamclinicaldl.dfs.core.windows.net/{t}/'
    """)
    print(f"Registered: databricks_clinical_trial.silver.{t}")

print("\nAll 18 Silver tables registered in Unity Catalog.")
```

## Step 6 Ã¢â‚¬â€ Gold Layer: Business Analytics Tables
### Architecture
Gold tables join cleaned Silver entities into analytics-ready views. Each table serves a specific business question. All tables are external Delta tables registered in Unity Catalog and stored in the Gold container for downstream Synapse Analytics consumption.

| Table | Grain | Source Tables | Business Purpose |
|---|---|---|---|
| `patient_study_summary` | 1 row per enrollment | patients, enrollments, clinical_studies, study_arms, research_sites | Patient-level clinical overview |
| `lab_summary` | 1 row per lab result | lab_results, patients, clinical_visits, enrollments | Lab analytics with toxicity risk classification |
| `safety_adverse_events` | 1 row per adverse event | adverse_events, patients, enrollments | Safety monitoring and AE risk categorization |
| `visit_summary` | 1 row per clinical visit | clinical_visits, enrollments, patients | Visit compliance and timing analysis |

### Gold Table 1 Ã¢â‚¬â€ `patient_study_summary`
One row per patient enrollment. Joins patient demographics with study, arm, and site context. Used for enrollment reporting and patient-level clinical overviews.

**Expected rows:** 200

```python
%sql
CREATE OR REPLACE TABLE databricks_clinical_trial.gold.patient_study_summary
USING DELTA
LOCATION 'abfss://gold@hishamclinicaldl.dfs.core.windows.net/patient_study_summary/'
AS
SELECT
    p.patient_id,
    p.patient_code,
    p.gender,
    p.race,
    p.ethnicity,
    p.date_of_birth,
    floor(datediff(current_date(), p.date_of_birth) / 365.25) AS age_years,

    e.enrollment_id,
    e.enrollment_date,
    e.status                AS enrollment_status,

    s.study_id,
    s.study_code,
    s.study_title,
    s.therapeutic_area,
    s.phase,

    a.arm_id,
    a.arm_name,

    r.site_id,
    r.site_name
FROM databricks_clinical_trial.silver.patients          p
INNER JOIN databricks_clinical_trial.silver.enrollments e
    ON p.patient_id = e.patient_id
INNER JOIN databricks_clinical_trial.silver.clinical_studies s
    ON e.study_id = s.study_id
LEFT JOIN  databricks_clinical_trial.silver.study_arms  a
    ON e.arm_id = a.arm_id
LEFT JOIN  databricks_clinical_trial.silver.research_sites r
    ON e.site_id = r.site_id;

SELECT COUNT(*) AS row_count FROM databricks_clinical_trial.gold.patient_study_summary;
```

### Gold Table 2 Ã¢â‚¬â€ `lab_summary`
One row per lab result. Joins lab data with patient demographics and study context via the visitÃ¢â€ â€™enrollment path (correct join path to avoid row multiplication). Adds lab status and toxicity risk classification.

**Expected rows:** 7715

```python
%sql
CREATE OR REPLACE TABLE databricks_clinical_trial.gold.lab_summary
USING DELTA
LOCATION 'abfss://gold@hishamclinicaldl.dfs.core.windows.net/lab_summary/'
AS
SELECT
    l.lab_result_id,
    l.visit_id,
    l.patient_id,

    p.patient_code,
    p.gender,
    floor(datediff(current_date(), p.date_of_birth) / 365.25) AS age_years,

    v.enrollment_id,
    e.study_id,
    e.site_id,

    l.lab_test_name,
    l.lab_test_code,
    l.lab_panel,
    l.result_value,
    l.result_unit,
    l.normal_range_low,
    l.normal_range_high,
    l.is_abnormal,
    l.toxicity_grade,

    CASE WHEN l.is_abnormal = true THEN 'Abnormal' ELSE 'Normal' END AS lab_status,

    CASE
        WHEN l.toxicity_grade >= 3 THEN 'High Risk'
        WHEN l.toxicity_grade = 2  THEN 'Moderate Risk'
        ELSE 'Low Risk'
    END AS toxicity_risk_category,

    l.collection_datetime,
    l.result_datetime
FROM databricks_clinical_trial.silver.lab_results       l
LEFT JOIN databricks_clinical_trial.silver.patients     p
    ON l.patient_id = p.patient_id
LEFT JOIN databricks_clinical_trial.silver.clinical_visits v
    ON l.visit_id = v.visit_id
LEFT JOIN databricks_clinical_trial.silver.enrollments  e
    ON v.enrollment_id = e.enrollment_id;

SELECT COUNT(*) AS row_count FROM databricks_clinical_trial.gold.lab_summary;
```

### Gold Table 3 Ã¢â‚¬â€ `safety_adverse_events`
One row per adverse event. Joins AE data with patient demographics and enrollment context. Adds AE risk category, resolution status, and duration in days for safety monitoring.

**Expected rows:** 245

```python
%sql
CREATE OR REPLACE TABLE databricks_clinical_trial.gold.safety_adverse_events
USING DELTA
LOCATION 'abfss://gold@hishamclinicaldl.dfs.core.windows.net/safety_adverse_events/'
AS
SELECT
    ae.ae_id,
    ae.patient_id,
    ae.visit_id,
    ae.enrollment_id,
    ae.reporter_investigator_id,

    p.patient_code,
    p.gender,
    floor(datediff(current_date(), p.date_of_birth) / 365.25) AS age_years,

    e.study_id,
    e.site_id,

    ae.ae_term,
    ae.meddra_pt_code,
    ae.severity,
    ae.ctcae_grade,
    ae.onset_date,
    ae.resolution_date,
    ae.outcome,
    ae.is_serious,
    ae.sae_category,
    ae.causality,
    ae.action_taken,
    ae.is_expected,

    CASE
        WHEN ae.is_serious = true    THEN 'Serious AE'
        WHEN ae.ctcae_grade >= 3     THEN 'High Grade AE'
        WHEN ae.ctcae_grade = 2      THEN 'Moderate AE'
        ELSE 'Low Grade AE'
    END AS ae_risk_category,

    CASE
        WHEN ae.resolution_date IS NULL THEN 'Ongoing'
        ELSE 'Resolved'
    END AS ae_status,

    datediff(ae.resolution_date, ae.onset_date) AS ae_duration_days,

    ae.created_at,
    ae.updated_at
FROM databricks_clinical_trial.silver.adverse_events    ae
LEFT JOIN databricks_clinical_trial.silver.patients     p
    ON ae.patient_id = p.patient_id
LEFT JOIN databricks_clinical_trial.silver.enrollments  e
    ON ae.enrollment_id = e.enrollment_id;

SELECT COUNT(*) AS row_count FROM databricks_clinical_trial.gold.safety_adverse_events;
```

### Gold Table 4 Ã¢â‚¬â€ `visit_summary`
One row per clinical visit. Joins visit scheduling data with patient and study context. Adds visit timing status (On Time / Late / Early / Not Completed) and days deviation from scheduled date.

**Expected rows:** 1108

```python
%sql
CREATE OR REPLACE TABLE databricks_clinical_trial.gold.visit_summary
USING DELTA
LOCATION 'abfss://gold@hishamclinicaldl.dfs.core.windows.net/visit_summary/'
AS
SELECT
    v.visit_id,
    v.enrollment_id,
    v.investigator_id,
    v.visit_number,
    v.visit_name,
    v.visit_type,

    e.patient_id,
    p.patient_code,
    e.study_id,
    e.site_id,

    v.scheduled_date,
    v.actual_visit_date,
    v.visit_window_start,
    v.visit_window_end,
    v.visit_status,

    CASE
        WHEN v.actual_visit_date IS NULL                                           THEN 'Not Completed'
        WHEN v.actual_visit_date BETWEEN v.visit_window_start AND v.visit_window_end THEN 'On Time'
        WHEN v.actual_visit_date < v.visit_window_start                            THEN 'Early'
        WHEN v.actual_visit_date > v.visit_window_end                              THEN 'Late'
        ELSE 'Unknown'
    END AS visit_timing_status,

    datediff(v.actual_visit_date, v.scheduled_date) AS days_from_scheduled,

    v.notes,
    v.created_at,
    v.updated_at
FROM databricks_clinical_trial.silver.clinical_visits   v
LEFT JOIN databricks_clinical_trial.silver.enrollments  e
    ON v.enrollment_id = e.enrollment_id
LEFT JOIN databricks_clinical_trial.silver.patients     p
    ON e.patient_id = p.patient_id;

SELECT COUNT(*) AS row_count FROM databricks_clinical_trial.gold.visit_summary;
```

## Step 7 Ã¢â‚¬â€ Final Verification
Confirms all four Gold tables exist in Unity Catalog and returns the correct row counts. Run this after any re-execution to validate end-to-end pipeline integrity.

```python
%sql
SELECT 'patient_study_summary'  AS gold_table, COUNT(*) AS row_count FROM databricks_clinical_trial.gold.patient_study_summary
UNION ALL
SELECT 'lab_summary',                           COUNT(*) FROM databricks_clinical_trial.gold.lab_summary
UNION ALL
SELECT 'safety_adverse_events',                 COUNT(*) FROM databricks_clinical_trial.gold.safety_adverse_events
UNION ALL
SELECT 'visit_summary',                         COUNT(*) FROM databricks_clinical_trial.gold.visit_summary
ORDER BY gold_table;
```
